````python
import numpy as np
import matplotlib.pyplot as plt

# --- È°åÁõÆÂ∑≤Êèê‰æõÁöÑÁ®ãÂºèÁ¢º ---
# Âõ∫ÂÆö‰∫ÇÊï∏Á®ÆÂ≠ê‰ª•‰æøÁµêÊûúÂèØÈáçÁèæ
np.random.seed(768)
n = 200

# --- Ë´ãÂú®ÈÄôË£°Âä†ÂÖ•ÊÇ®ÁöÑÁ®ãÂºèÁ¢º ---

# 1. Ê†πÊìö„ÄåÂèÉËÄÉÁ®ãÂºè„ÄçÁîüÊàê x, y, size_var
# (ÊàëÂÄëÁõ¥Êé•ÂæûÈ°åÁõÆË™™Êòé‰∏≠Ë§áË£ΩÈÅé‰æÜ)
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))

# 2. ÁÇ∫‰∫ÜË¶ñË¶∫ÊïàÊûúÔºåÂ∞ç size_var ÂÅöÈÅ©Â∫¶Á∏ÆÊîæ
#    (matplotlib ÁöÑ s ÂèÉÊï∏ÊòØÈù¢Á©çÔºåÊâÄ‰ª•ÊàëÂÄëÁõ¥Êé•Áî® size_var ÊïàÊûúÂèØËÉΩ‰∏çÂ•Ω)
#    (ÊàëÂÄëÂ∞á size_var ‰πò‰ª•‰∏ÄÂÄãÂÄçÊï∏Ôºå‰æãÂ¶Ç 10Ôºå‰ΩøÂÖ∂Ë¶ñË¶∫‰∏äÊõ¥ÊòéÈ°Ø)
size_scale = size_var * 10 
# (Ê≥®ÊÑèÔºöÊÇ®‰πüÂèØ‰ª•ÂòóË©¶‰∏çÂêåÁöÑÁ∏ÆÊîæÂÄçÊï∏Ôºå‰æãÂ¶Ç 5 Êàñ 15ÔºåÁúãÁúãÊïàÊûú)
# (ÊàñËÄÖÂ¶ÇÊèêÁ§∫ÊâÄË™™ÔºåÁõ¥Êé•‰ΩøÁî® size_var ‰πüÂèØ‰ª•Ôºö size_scale = size_var)

# 3. Áπ™Ë£ΩÊï£‰ΩàÂúñ (plt.scatter)
#    x Ëª∏: x
#    y Ëª∏: y
#    È°èËâ≤ (c): Ê†πÊìö y ÂÄºËÆäÂåñ
#    Â§ßÂ∞è (s): Ê†πÊìöÁ∏ÆÊîæÂæåÁöÑ size_scale ËÆäÂåñ
#    cmap: ÊåáÂÆö‰∏ÄÂÄãÈ°èËâ≤Â∞çÊò†Ë°® (colormap)Ôºå‰æãÂ¶Ç 'viridis' Êàñ 'plasma'
#    alpha: Ë®≠ÂÆöÈÄèÊòéÂ∫¶ (‰æãÂ¶Ç 0.6)ÔºåÈÅøÂÖçÈªûÈáçÁñäÊôÇÁúã‰∏çÊ∏ÖÊ•ö
scatter = plt.scatter(x, y, c=y, s=size_scale, cmap='viridis', alpha=0.6, edgecolor='k')

# 4. Ê†πÊìöË¶ÅÊ±ÇË®≠ÂÆöÊ®ôÈ°å
plt.title("Scatter with Colormap")

# 5. Ê†πÊìöË¶ÅÊ±ÇË®≠ÂÆöÈ°èËâ≤Ê¢ù (Colorbar)
#    plt.colorbar() ÊúÉËá™ÂãïÊäìÂèñ plt.scatter ÁöÑÈ°èËâ≤Â∞çÊò†
#    ‰∏¶ÁÇ∫ÂÖ∂Ë®≠ÂÆöÊ®ôÁ±§
cbar = plt.colorbar(scatter)
cbar.set_label("y-values")

# (x, y Ëª∏Ê®ôÁ±§È°åÁõÆÈõñÁÑ∂Ê≤íË¶ÅÊ±ÇÔºå‰ΩÜÂä†‰∏äÂéªÊõ¥ÂÆåÊï¥)
plt.xlabel("x-values")
plt.ylabel("y-values")

# --- Êñ∞Â¢ûÁµêÊùü ---

plt.show()
````
