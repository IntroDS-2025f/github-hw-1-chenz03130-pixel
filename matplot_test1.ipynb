````python
import numpy as np
import matplotlib.pyplot as plt

# --- 題目已提供的程式碼 ---
# 固定亂數種子以便結果可重現
np.random.seed(768)
n = 200

# --- 請在這裡加入您的程式碼 ---

# 1. 根據「參考程式」生成 x, y, size_var
# (我們直接從題目說明中複製過來)
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))

# 2. 為了視覺效果，對 size_var 做適度縮放
#    (matplotlib 的 s 參數是面積，所以我們直接用 size_var 效果可能不好)
#    (我們將 size_var 乘以一個倍數，例如 10，使其視覺上更明顯)
size_scale = size_var * 10 
# (注意：您也可以嘗試不同的縮放倍數，例如 5 或 15，看看效果)
# (或者如提示所說，直接使用 size_var 也可以： size_scale = size_var)

# 3. 繪製散佈圖 (plt.scatter)
#    x 軸: x
#    y 軸: y
#    顏色 (c): 根據 y 值變化
#    大小 (s): 根據縮放後的 size_scale 變化
#    cmap: 指定一個顏色對映表 (colormap)，例如 'viridis' 或 'plasma'
#    alpha: 設定透明度 (例如 0.6)，避免點重疊時看不清楚
scatter = plt.scatter(x, y, c=y, s=size_scale, cmap='viridis', alpha=0.6, edgecolor='k')

# 4. 根據要求設定標題
plt.title("Scatter with Colormap")

# 5. 根據要求設定顏色條 (Colorbar)
#    plt.colorbar() 會自動抓取 plt.scatter 的顏色對映
#    並為其設定標籤
cbar = plt.colorbar(scatter)
cbar.set_label("y-values")

# (x, y 軸標籤題目雖然沒要求，但加上去更完整)
plt.xlabel("x-values")
plt.ylabel("y-values")

# --- 新增結束 ---

plt.show()
````
